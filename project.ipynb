```python
import json
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Load JSON data
json_path = '2017_6217East/2017_6217EastDJI_0130.json'
with open(json_path) as f:
    data = json.load(f)

# Load sample image
sample_id = data['sample_id']
image_path = f'2017_6217East/2017_6217East{sample_id}.JPG'
image_bgr = cv2.imread(image_path)
image = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Plot boundary over image
fig, ax = plt.subplots(figsize=(12, 10))
ax.axis('off')
ax.imshow(image)
for plant in data['labels']:
    polygon_flat = plant['segment']
    polygon_points = np.reshape(polygon_flat, (len(polygon_flat) // 2, 2))
    ax.add_patch(Polygon(polygon_points, fill=False, edgecolor='red', linewidth=2))
```

# Helper Functions

```python
def plot_image(image):
    fig, ax = plt.subplots(figsize=(12, 10))
    ax.imshow(cv2.cvtColor(image, cv.COLOR_BGR2RGB))
    ax.axis('off')
    return fig, ax
```
